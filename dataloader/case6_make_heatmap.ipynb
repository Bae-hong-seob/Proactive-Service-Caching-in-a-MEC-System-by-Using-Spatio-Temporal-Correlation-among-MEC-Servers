```python
import pandas as pd

number_of_sample_users = 5000

file_name = 'rand_grad5_2k.txt'

try:
    ex_rating_df = pd.read_table('C:\\content_caching\\'+file_name, sep=' ', header=None)
except:
    ex_rating_df = pd.read_csv('/content/drive/MyDrive/jaesungpark_lab/Paper_review/Content_Caching/'+file_name)

ex_rating_df
```

```python
try:
    ex_rating_df = ex_rating_df.drop(128, axis=1)
except:
    print('No index column')

ex_rating_df
```

```python
# Normalize each row(=each mecs)
ex_rating_df = ex_rating_df.div(ex_rating_df.max(axis=1), axis=0)
ex_rating_df
```

```python
import numpy as np

image_array = np.array(ex_rating_df)
image_array.shape
```

```python
number_of_mecs = 16
number_of_images = image_array.shape[0] // number_of_mecs

print('number of mecs : ', number_of_mecs)
print('number of images : ', number_of_images)
```

```python
import os

save_path = './'+file_name[:-4] # 확장자 제외

try:
    os.mkdir(save_path)
except:
    print('[Errno 17] File exists')
```

```python
from PIL import Image


for i in range(number_of_images):
    print(i,' 번째 이미지 생성중 ...')
    image = image_array[i:i+number_of_mecs, :]
    print(image.shape)    # Create a PIL image from the array
    #print(image)

    image = Image.fromarray((255 * image).astype(np.uint8), mode='L')

    file_name = '/train'+str(i)+'.png'
    image.save(save_path+file_name)
```
